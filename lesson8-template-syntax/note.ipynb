# 什么是响应性reactivity？

其实就是这个函数

```js
import {ref} from 'vue'
const name = ref('doctor')
```

可以通过const赋予变量各种类型的值，比如文本、数字、字典、布尔值等

注意，你可以通过

some_ref.value= "xxxx"

直接修改这个ref的值，案例见Header.vue

# 模板语法：文字插值 template syntax text interpolation

以Header.vue为例

title是一个ref对象，这个ref有很多属性，比如值value

你可以直接改变title这个ref的值，新的变量会覆盖旧的变量，vuejs会自动解析最新的ref.value

接下来就可以动态修改这些值，比如把后端的内容传到这些ref对象里，从而即时渲染页面发生变化
