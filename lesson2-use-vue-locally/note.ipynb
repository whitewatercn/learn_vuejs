# 解读一下Vue.js模板创建好以后，生成的文件有什么用

```python
❯ tree -L 2
.
├── index.html # 入口，一般不用动
├── node_modules # 一堆安装的依赖包
├── public # 打算暴露给外部访问的资源
│   └── favicon.ico
├── src
│   ├── App.vue # vue项目的核心 ！！！
│   └── main.js # 把 App.vue 挂载到index.html上去
# 下面是配置文件
├── jsconfig.json # 构建项目时要编译哪些文件，一般是编译src文件夹下所有内容，排除node_modules、dist下所有内容
├── package.json # 记录依赖包版本等信息
├── package-lock.json
└── vite.config.js # vite的配置文件，vite是一个构建工具，在本项目中，暂时用不到 vue-devtools 插件，先注释掉
```
